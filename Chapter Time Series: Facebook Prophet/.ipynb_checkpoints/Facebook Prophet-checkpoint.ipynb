```python
import numpy as np
import pandas as pd
from prophet import Prophet

import plotly.express as px
import plotly.graph_objs as go
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

## Load Data & Preprocessing

```python
df = pd.read_csv('electricityConsumptionAndProductioction.csv')
```

```python
df['Date'] = pd.to_datetime(df['DateTime']).dt.date
```

```python
df_2 = df.groupby(['Date'],as_index= False).sum()
```

```python
fig = px.line(
    df_2,
    x = 'Date',
    y = 'Consumption',
    title = 'Daily Consumption',
    width = 1000,
    height = 600)
fig.show()
```

## Facebook Prophet

```python
# add summer season with different weekly pattern
def is_summer_season(ds):
    date = pd.to_datetime(ds)
    return ((date.month >= 6) & (date.month<=8))

df_2['on_season'] = df_2['Date'].apply(is_summer_season)
df_2['off_season'] = ~df_2['Date'].apply(is_summer_season)

# preparing for the training data
cap = df_2.Consumption.max()
floor = 0
df_2['cap'] = cap
df_2['floor'] = floor

df_train = df_2[['Date','Consumption','on_season','off_season','cap','floor']]
df_train['y'] = df_train['Consumption']
df_train['ds'] = df_train['Date']

#convert to log
df_train['y'] = df_train['y'] + 1
df_train['y'] = np.log(df_train['y'])

# add covid period as holiday to skip
lockdowns = pd.DataFrame([
    {
        'holiday':'covid_year1',
        'ds': pd.to_datetime('2020-03-17'),
        'lower_window':0,
        'ds_upper':pd.to_datetime('2021-03-17')
    }
])

lockdowns['upper_window'] = (lockdowns['ds_upper']-lockdowns['ds']).dt.days
```

```python
m = Prophet(weekly_seasonality = False,changepoint_range = 0.2,changepoint_prior_scale = 0.1,
           seasonality_prior_scale=0.01,holidays = lockdowns)
m.add_country_holidays(country_name = 'US')
m.add_seasonality(name = 'weekly_on_season',period = 7,fourier_order=4,condition_name='on_season')
m.add_seasonality(name = 'weekly_off_season',period = 7,fourier_order=4,condition_name='off_season')
model = m.fit(df_train)
```

```python
# training loss total
future = m.make_future_dataframe(periods=0)
future['cap'] = np.log(cap+1)
future['floor'] = floor
future['on_season'] = future['ds'].apply(is_summer_season)
future['off_season'] = ~future['ds'].apply(is_summer_season)
forecast = model.predict(future)
y_pred = list(np.exp(forecast.yhat)-1)
se = np.square(y_pred - df_train.Consumption)
mse = np.mean(se)
rmse = np.sqrt(mse)
rmse
```

```python
future_dates = model.make_future_dataframe(periods=700,freq = 'D')
future_dates['cap'] = np.log(cap+1)
future_dates['floor'] = floor
future_dates['on_season'] = future_dates['ds'].apply(is_summer_season)
future_dates['off_season'] = ~future_dates['ds'].apply(is_summer_season)
df_pred = model.predict(future_dates)
```

```python
df_plot = pd.DataFrame()
consumption = list(np.exp(df_pred.yhat)-1)
df_plot['Consumption'] = consumption
df_plot['Date'] = list(df_pred.ds)
```

```python
fig = px.line(
    df_plot,
    x = 'Date',
    y = 'Consumption',
    title = 'Prophet prediction',
    width = 1000,
    height = 600    
)

fig.add_trace(go.Scatter(x=df_2.Date,y=df_2.Consumption,mode='lines',
                        line = dict(color = 'green',
                                   width = 2,
                                   dash = 'dash'), name = 'Actual',showlegend=True))
fig.update_yaxes(matches = None)
fig['data'][0]['showlegend'] = True
fig['data'][0]['name']='Predicted'
fig.show()
```
